```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from nltk import pos_tag

lemmatizer = WordNetLemmatizer()

def lemmatize_verbs(text):
  words = word_tokenize(text)
  lemmas = []
  for word, tag in pos_tag(words):
    if tag.startswith('VB'):
      lemma = lemmatizer.lemmatize(word, pos='v')
    else:
      lemma = word
    lemmas.append(lemma)
  return ' '.join(lemmas)

lemmatized_text = lemmatize_verbs(text)
print(lemmatized_text)
```

```python
# Identify occurrences of the phrase "Maratha Empire" using spaCy's PhraseMatcher.

from spacy.matcher import PhraseMatcher

nlp = spacy.load('en_core_web_sm')
text = test

matcher = PhraseMatcher(nlp.vocab)
pattern = nlp("Maratha Empire")
matcher.add("MARATHA_EMPIRE", None, pattern)

doc = nlp(text)
matches = matcher(doc)

for match_id, start, end in matches:
    print(f"found at index: {start} - {end}, Phrase: '{doc[start:end]}'")
```

# 13- Keyword Extraction

Extract keywords using NLTK’s frequency distribution (common nouns and verbs).

Use spaCy to identify and list unique keywords from the text (filter out stop words).

```python
# Extract keywords using NLTK’s frequency distribution (common nouns and verbs).

import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.probability import FreqDist


# Tokenize the text
tokens = nltk.word_tokenize(test)

stop_words = set(stopwords.words('english'))
tokens = [token.lower() for token in tokens if token.isalpha() and token not in stop_words]

lemmatizer = WordNetLemmatizer()
tokens = [lemmatizer.lemmatize(token) for token in tokens]

freq_dist = FreqDist(tokens)

print("Top 10 Most Common Nouns and Verbs:")
print("*"*40)
for word, freq in freq_dist.most_common(10):
    print(f"{word}: {freq}")
print("*"*40)    
```

# 15 Sentiment Polarity Distribution

Calculate and plot the sentiment polarity distribution for each sentence in the text using TextBlob.

Identify and display the most positive and most negative sentences based on sentiment scores.

```python
import matplotlib.pyplot as plt

sentences = test.split('. ')

# Calculating the sentiment polarity for each sentence
sentiments = [TextBlob(sentence).sentiment.polarity for sentence in sentences]

# Plot the sentiment polarity distribution
plt.figure(figsize=(10, 6))
plt.plot(sentiments)
plt.title('Sentiment Polarity Distribution')
plt.xlabel('Sentence Index')
plt.ylabel('Sentiment Polarity')
plt.show()
```

```python
# Identify and display the most positive and most negative sentences
max_sentiment = max(sentiments)
min_sentiment = min(sentiments)

max_sentiment_index = sentiments.index(max_sentiment)
min_sentiment_index = sentiments.index(min_sentiment)

print(f"Most Positive Sentence: {sentences[max_sentiment_index]} (Sentiment Score: {max_sentiment})")
print(f"Most Negative Sentence: {sentences[min_sentiment_index]} (Sentiment Score: {min_sentiment})")
```

# 19- Extract Subject-Verb-Object Triples

Use spaCy to extract subject-verb-object triples from the sentences.

Identify and list sentences where "Shivaji" is the subject and a verb follows, using dependency parsing.

```python
# Extract subject-verb-object triples
for sentence in doc.sents:
    for token in sentence:
        if token.dep_ == 'nsubj':
            subject = token.text
            verb = token.head.text
            object = None
            for child in token.head.children:
                if child.dep_ == 'dobj':
                    object = child.text
            if object:
                print(f"Subject: {subject}, Verb: {verb}, Object: {object}")

```
