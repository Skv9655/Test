1. Write a function `fibonacci(n)` that generates the first `n` numbers in the Fibonacci sequence and returns




````python
def fib(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]

    fib_seq = [0, 1]
    for i in range(2, n):
        value = fib_seq[-1] + fib_seq[-2]
        fib_seq.append(value)

    return fib_seq
````

````python
fib(10)
````

2. Write a function `is_prime(num)` that checks if a given number is a prime number. Return `True` if the nu

````python
num = int(input("Enter the number to check Prime or not "))

if num > 1:
    for i in range(2, num):
        if (num % i) == 0:
            print(False)
            break
    else:
        print(True)
````
